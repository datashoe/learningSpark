```python
# 构造一个线性回归的数据集
import numpy as np
x_data = np.random.rand(1000) 
y_data = x_data * 2.0 + 1.5

# 利用tensorflow中的梯度下降法训练该线性回归
import tensorflow as tf

# 定义线性回归的两个变量：系数和常数项
k = tf.Variable(0.0)
b = tf.Variable(0.0)
y = x_data * k + b

# 梯度下降需要：学习率、损失函数 --梯度下降会自动根据损失函数求导，这一点很方便
loss = tf.reduce_mean(tf.square(y_data - y))
optimizer = tf.train.GradientDescentOptimizer(0.2)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()

# 运行梯度下降4000次并打印每次的结果和对应的损失函数
ckpt_file_path = "E://company/documents/miner develop/test_ckpt/"
with tf.Session() as session:
    session.run(init)
    for step in range(4000):
        session.run(train)
        print(session.run([k, b, loss]))
    tf.train.Saver().save(session, ckpt_file_path)
```
